```python
import requests
import json
import os

def descargar_json_por_scope_id(base_url, scope_id):
    """Descarga el contenido JSON de un scopeId específico."""
    url = base_url.format(scope_id)
    response = requests.get(url)
    response.raise_for_status()
    return response.json()

def guardar_json_en_archivo(data, filename):
    """Guarda el contenido JSON en un archivo."""
    with open(filename, 'w', encoding='utf-8') as file:
        json.dump(data, file, ensure_ascii=False, indent=4)

def main():
    # URL base
    base_url = "https://paso.resultados.gob.ar/backend-difu/scope/data/getScopeData/{}/1/1"
    
    # Descargar el contenido JSON inicial
    scope_id_inicial = "00000000000000000000000b"
    data_inicial = descargar_json_por_scope_id(base_url, scope_id_inicial)
    
    # Guardar el contenido inicial en el archivo PASO_Nacionales.json
    guardar_json_en_archivo(data_inicial, "PASO_Nacionales.json")
    print("Archivo PASO_Nacionales.json guardado con éxito.")
    
    # Extraer todos los scopeId y name del JSON
    for item in data_inicial.get("mapa", []):
        for scope in item.get("scopes", []):
            scope_id = scope.get("scopeId")
            name = scope.get("name")
            
            # Descargar el contenido JSON de cada scopeId
            data_scope = descargar_json_por_scope_id(base_url, scope_id)
            
            # Guardar el contenido en un archivo
            filename = f"PASO_{name}.json"
            guardar_json_en_archivo(data_scope, filename)
            print(f"Archivo {filename} guardado con éxito.")

if __name__ == "__main__":
    main()
```
