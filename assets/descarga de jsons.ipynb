```python
import requests
import os
import json

def descargar_y_guardar(url, path):
    """Descargar contenido de URL y guardarlo en path"""
    response = requests.get(url)
    response.raise_for_status()  # Asegurarse de que la respuesta es 200 OK
    
    # Obtener el contenido del JSON
    data = response.json()

    # Escribir el archivo GeoJSON
    with open(path, 'w', encoding='utf-8') as file:
        json.dump(data, file, ensure_ascii=False, indent=4)

# URL del archivo principal
url_principal = "https://paso.resultados.gob.ar/assets/mapas/1.json"

# Descargar el archivo principal
response = requests.get(url_principal)
response.raise_for_status()

# Obtener el contenido del archivo principal
data_principal = response.json()

# Asegurarse de que la carpeta MAPA existe o crearla si no existe
if not os.path.exists("MAPAS"):
    os.makedirs("MAPAS")

# Guardar el archivo principal como MAPA_Nacionales.geojson
path_principal = os.path.join("MAPAS", "MAPA_Nacionales.geojson")
with open(path_principal, 'w', encoding='utf-8') as file:
    json.dump(data_principal, file, ensure_ascii=False, indent=4)
print(f"Archivo MAPA_Nacionales.geojson guardado con éxito.")

# Extraer los valores 'name' de cada 'feature'
nombres = [feature['properties']['name'] for feature in data_principal['features']]

# Base del enlace para descargar cada archivo individual
base_url = "https://paso.resultados.gob.ar/assets/mapas/{}.json"

# Iterar sobre cada 'name' y descargar el archivo correspondiente
for nombre in nombres:
    url_individual = base_url.format(nombre)
    path = os.path.join("MAPAS", f"{nombre}.geojson")
    descargar_y_guardar(url_individual, path)
    print(f"Archivo {nombre}.geojson guardado con éxito.")

print("Todos los archivos han sido descargados y guardados.")
```
